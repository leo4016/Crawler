# 抓網址
### 單tread 10 分鐘

```python
import requests
from bs4 import BeautifulSoup
import json
import re
from datetime import datetime, timedelta, date

DOMAIN = "http://news.ltn.com.tw"
dates = []
d_format = "%Y%m%d"
date = datetime.strptime("20140101", d_format)
delta = timedelta(days=1)
today = datetime.strptime(date.today().strftime(d_format), d_format)
while date <= today:
    dates.append(date.strftime("%Y%m%d"))
    date += delta

url_list = []
DOMAIN = "http://news.ltn.com.tw/"
for pg in dates:
    URL = 'http://news.ltn.com.tw/list/newspaper/politics/{}'
    resp = requests.get(URL.format(pg))
    soup = BeautifulSoup(resp.text,'html5lib')
    for item in range(0,len(soup.select('li > a.tit'))):#自動抓筆數,
        urls = DOMAIN+soup.select('li > a.tit')[item]['href']
        news_list.append(urls)
with open("./liberty_news.text", 'w') as f:
    f.write(json.dumps(url_list))
    f.close()
url_list
```

# 爬內容
###  (單thread )

```python
raw_url = json.load(open('./liberty_news.text'))#load for file, loads for string

news_list = []
for record in range(0,len(raw_url)):
    URLC = raw_url[record]
    
    respc = requests.get(URLC)
    soup = BeautifulSoup(respc.text,'html5lib')
   
    news = {}
    news["title"] = soup.select("div.articlebody > h1")[0].text.replace("\n", "").replace("\t", "")
    news["datetime"] = soup.select("div.text > span")[0].text
    content = ''
    for i in range(0,len(soup.select(' div.articlebody > div.text > p'))):
        content += soup.select(' div.articlebody > div.text > p')[i].text
    news["content"] = content 
    news["media"] = "liberty times net"
    news["category"] = soup.select('section > div.boxTitle > a')[2].text
    news["url"] = URLC
    news["reporter"] = re.findall('記者(.+)／.+報導.+',soup.select(' div.articlebody > div.text > p')[0].text)
    news["location"] = re.findall('記者.+／(.+)報導.+',soup.select(' div.articlebody > div.text > p')[0].text)
   
    news_list.append(news)
with open("./liberty_news.csv", 'w') as f:
    f.write(json.dumps(news_list, ensure_ascii=False))
    f.close()
```
