```python
import requests
from bs4 import BeautifulSoup
#URL = 'https://world.taobao.com/search/json.htm?navigator=all&cat=50000147&_ksTS=1510562051490_300&spm=a213z.1224559.20141209.1&json=on&cna=CD9NEiELYEMCAT3kpldgW9rm&_input_charset=utf-8&s=0&q=%E4%B9%A6&callback=__jsonp_cb&abtest=_AB-LR517-LR854-LR895-PR517-PR854-PR895&nid=&type=&uniqpid='
#URL 為javascript內的URL( https://world.taobao.com/search/json.htm?cat=50.............)
URL = 'https://world.taobao.com/search/json.htm?navigator=all&cat=50000147&_ksTS=1510562051490_300&spm=a213z.1224559.20141209.1&json=on&cna=CD9NEiELYEMCAT3kpldgW9rm&_input_charset=utf-8&s=0&q=%E4%B9%A6&callback=__jsonp_cb&abtest=_AB-LR517-LR854-LR895-PR517-PR854-PR895&nid=&type=&uniqpid='
resp = requests.get(URL)
resp.text
```

```python
import json
import re
m = re.search('\t if\(window.__jsonp_cb\)\{__jsonp_cb\((.*?)\)\}',resp.text)
jd = json.loads(m.group(1))
jd
```

```python
with open('a.json','w') as f:
    f.write(json.dumps(jd))
```
