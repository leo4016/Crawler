```python
import requests
import json
import re
from bs4 import BeautifulSoup
URL = 'http://www.etwarm.com.tw/object_list'
resp = requests.get(URL)
soup = BeautifulSoup(resp.text,'html5lib')
house = soup.select('li.obj_item > div.obj_info > h3 > a')
for page in range(1,11):
    list = []
    for x in range(0,len(house)):
        title = soup.select('li.obj_item > div.obj_info > h3 > a')[x].text
        price = soup.select('div.price')[x].text
        content = (title+price+'萬')
        #content = ('標的物'+title+'價錢'+price+'萬')
       # print('標的物'+title+'價錢'+price+'萬')
        list.append(content)
        print(list)
    with open('./東森房屋%s.txt'%page, 'w') as f:
        f.write(json.dumps(list, ensure_ascii=False))
        f.close()
    
    
```
