```python
def get_detail_page(detail_url):
    """
        input <str>: book.com.tw detail page url of a book
        
        return <dict>: include author, intro,press,title
    """
    resp1 = requests.get(detail_url)
    r = BeautifulSoup(resp1.text, 'html5lib')
    articles = {}
    
    if r.find("div",{"style":"text-align: center;"}):
        articles["introduction"] = r.find("div",{"style":"text-align: center;"}).get_text()
    else:
        articles['introduction'] = ''
        
    if  r.find("h1",{"itemprop":"name"}):
        articles["title"] = r.find("h1",{"itemprop":"name"}).get_text()
    else:
        articles['title'] = ''
        
    if  r.find("a",href=re.compile("http://search.books.com.tw/exep/.+")):
        articles["author"] = r.find("a",href=re.compile("http://search.books.com.tw/exep/.+")).get_text()
    else:
        articles["author"] = ''
        
    if  r.find("span",{"itemprop":"brand"}):
        articles["press"] = r.find("span",{"itemprop":"brand"}).get_text()
    else:
        articles["press"] = ''
        
        
            
        
        
    return articles
    
    
```
